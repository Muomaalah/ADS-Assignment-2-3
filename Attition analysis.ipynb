```python
Job_Dist_Attr = df.pivot_table(values=["DistanceFromHome"], index=["Attrition", "JobRole"], aggfunc=np.mean).reset_index()
```

```python
Job_Dist_Attr
```

```python
import seaborn as sns

plt.style.use('seaborn-bright')

plt.figure(figsize=(15,8))

plt.xticks(rotation=90)

sns.barplot(data = Job_Dist_Attr, x= 'JobRole', y= 'DistanceFromHome', hue='Attrition')
```

```python
income_edu = df.pivot_table(values=["MonthlyIncome"], index=["Attrition", "EducationField"], aggfunc=np.mean).reset_index()
```

```python
income_edu
```

```python
import seaborn as sns

plt.style.use('seaborn-bright')

plt.figure(figsize=(15,8))

plt.xticks(rotation=90)

sns.barplot(data = income_edu, x= 'EducationField', y= 'MonthlyIncome', hue='Attrition')
```
